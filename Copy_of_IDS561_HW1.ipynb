
Data Cleaning 

```
# function to stem using designated stemmer

def stem_tokens(tokens, stemmer):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    return stemmed
```

```
# function to preprocess: tokenize, make lower case and remove non-alphanumeric characters, 
# and apply designated stemmer

def Clean_(document):
  doc_list = []
  for line in document:
    # tokenize string
    tokens = nltk.word_tokenize(line)
  
    # rmake lower case and remove non-alphanumeric characters
    words = [word.lower() for word in tokens if word.isalpha()]

    # stem with PorterStemmer
    stems = stem_tokens(words, PorterStemmer())
    stems = " ".join(stems)
    if stems != "" : doc_list.append(stems) 

  return doc_list
```

```
pp_clean = Clean_(pp_txt)
print(pp_clean[:3])
```

Split

```
def Split_(document):
  Part1 = document[:5000]
  Part2 = document[5000:]

  return [Part1, Part2]
```

```
pp_split_1 = Split_(pp_clean)[0]
pp_split_2 = Split_(pp_clean)[1]

print(pp_split_1)
```

Mappers

```
# mapper function 

def Mapper_(document):
  tf = {}
  for i in document:
    for word in i.split(" "): 
      if word in tf.keys():
        tf[word].append(1)
      else:
        tf[word] = [1]
  return tf
```

```
# Mappers 1 and 2

Mapper1 = Mapper_(pp_split_1)
Mapper2 = Mapper_(pp_split_2)
```

Sorting

```
def Sort_(mapper1, mapper2):
  sort_document = []
  # mapper = {}
  # for k in mapper1.keys():
  #     if k in mapper2.keys():
  #       mapper

  sort_document.append(sorted(list(mapper1.keys())+ list(mapper2.keys())))
  sort_document.append(mapper1)
  sort_document.append(mapper2)
  return sort_document
```

```
Sorted= Sort_(Mapper1,Mapper2)
```

Partition

```
def Partition_(document):
  Part1 = list()
  Part2 = list()
  for i in document:
    if i[0]< "n":
      Part1.append(i)
    else:
      Part2.append(i)
  return (sorted(list(set(Part1))), sorted(list(set(Part2))))
```

```
Part1, Part2 = Partition_(Sorted[0])
```

Reducer

```
def Reducer_(Part):
  cnt_list = []
  for i in Part:
    s = 0
    if i in Mapper1.keys():
      s += sum(Mapper1[i])
    if i in Mapper2.keys():
      s += sum(Mapper2[i])
    cnt_list.append(s)

  a = pd.DataFrame({"Word":Part, "Frequency":cnt_list})
                     
  return a
```

```
Reducer1 = Reducer_(Part1)
Reducer2 = Reducer_(Part2)
```

```
Result = pd.concat([Reducer1, Reducer2])
print(Result)
```
